```python
# %load ../standard_import.txt
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
 
  

import datetime 
%matplotlib inline
plt.style.use('seaborn-white')

import os
import glob 

# use glob to get all the sim files in the folder
  
path = r"C:\Users\brianna.bai\Downloads\Project\Database\RawData" 
sim_files = glob.glob(path + "/*.xlsm")

df = pd.DataFrame()

for filename in sim_files:
     
    df1 = pd.read_excel(filename, sheet_name="Summary", skiprows=7, skipfooter=79, index_col=0)
    df1 = df1.dropna(axis=1).drop(columns=df1.columns[df1.eq(0).mean()>0.9]).T
    df2 = pd.read_excel(filename, sheet_name="Summary", skiprows=3, skipfooter=97, usecols="A:B", header=None)
    df1["Project_Name"]= df2.iloc[0][1]
    df1["Square_Footage"]= df2.iloc[1][1]
    df1= df1.reset_index(drop=True)
    #print(df1)
    df= df.reset_index(drop=True)
    print(df)
    df= pd.concat([df,df1], axis=1)

display(df) 

```
